# **Análisis exploratorio de los datos**

## **Descripción**

El proceso de Análisis Exploratorio de Datos (EDA) consiste en comprender cómo están estructurados los datos que queremos analizar.

Es un proceso de naturaleza investigativa, donde intentamos entender diversas características, como: los valores presentes en las columnas, los tipos de estructura de datos, verificar si son datos cualitativos o cuantitativos, si hay valores faltantes o inusuales.

Por lo tanto, en este momento, las preguntas sobre los datos siempre son bienvenidas. Estas guiarán todo el proceso de análisis, y a través de las herramientas disponibles, como nuestro querido Pandas, buscaremos respuestas.

Algunas preguntas que podemos hacer en este momento son:

1.   ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?
2.   ¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?"





```python
datos.head()
```

```python
datos['Valor'].mean()
```

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True).sort_values()
```

```python
df_tipo_precio = datos.groupby('Tipo')['Valor'].mean(numeric_only=True).sort_values()
df_tipo_precio.plot(kind='bar',figsize=(12,8), color='blue')
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/sala', 'Edificio Completo', 'Tienda/salón', 'Casa Comercial',
'Terreno Estandar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén',
'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio' , 'Industria']
```

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')['Valor'].mean(numeric_only=True).sort_values()
df_tipo_precio.plot(kind='bar',figsize=(12,8), color='blue')
```

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts()
```

```python
df.Tipo.value_counts(normalize=True)
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame()
porcentaje_tipo.plot(kind='bar',figsize=(12,8), color='blue', xlabel='Tipo', ylabel='Porcentaje')
```
